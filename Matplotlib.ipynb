Line Plot

```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
x
```

```python
y = 2 * x
y
```

```python
plt.plot(x, y)
plt.show()
```

Adding Title and Labels

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

Aesthetics and Multiple Lines

Changing Aesthetics

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
```

Adding Two Lines

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

Adding Subplots

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=2)
plt.show()
```

Matplotlib (Bar Plot)

```python
student = {"Bob":87,"Matt":56,"Sam":27}
names = list(student.keys())
values = list(student.values())
plt.bar(names, values)
plt.show()
```

```python
plt.bar(names, values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

Bar Plot (Horizontal)

```python
plt.barh(names,values,color='g')
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

Basic Scatter Plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a)
plt.show()
```

Changing Mark Aesthetics

```python
plt.scatter(x,a,marker="*",c="g",s=100)
plt.show()
```

Multiple Lines on One Plot

```python
b=[7,2,5,6,9,1,4,5,3]
plt.scatter(x,a,marker="*",c="g",s=100)
plt.scatter(x,b,marker=".",c="r",s=200)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="g",s=100)
plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="r",s=200)
plt.show()
```

Box Plot

```python
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]
data = list([one, two, three])
```

Making Plot

```python
plt.boxplot(data)
plt.show()
```

Violin Plot

```python
plt.violinplot(data, showmedians=True)
plt.show()
```

Matplotlib Pie Charts

```python
fruit = ['Apple', 'Orange', 'Mango', 'Guava']
quantity = [67, 34, 100, 29]
plt.pie(quantity, labels=fruit)
plt.show()
```

Adding Details

```python
plt.pie(quantity, labels=fruit, autopct='%0.1f%%', colors=['yellow', 'grey', 'blue', 'black'])
plt.show()
```

Doughnut Chart

```python
plt.pie(quantity, labels=fruit, radius=2)
plt.pie([1], colors=['w'], radius=1)
plt.show()
```
